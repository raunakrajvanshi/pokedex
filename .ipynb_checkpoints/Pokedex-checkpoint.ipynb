```python
import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Define the directory containing images
directory = "./images"

# Load the list of image file names
image_files = os.path.join(directory)
file_list = os.listdir(image_files)

# Display the list of image file names
print("List of image files:\n")
print(file_list)

# Plot sample images
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
first_six = file_list[0:6]

def plot_images():
    # Plot images
    row = 0
    col = 0
    for k in range(5):
        file_path = os.path.join(directory, first_six[k])
        img = plt.imread(file_path)
        axes[row,col].imshow(img)
        
        if k == 2:
            row += 1
            col = 0
        else:
            col += 1

    plt.tight_layout(pad=2)

plot_images()

# Load Pokémon data
pokemon_data = pd.read_csv("./pokemon.csv")

# Create a dictionary mapping Pokémon names to their types
pokemon_dict = dict(zip(pokemon_data["Name"], pokemon_data["Type1"]))

# Get unique Pokémon types
types = pokemon_data["Type1"].unique()

# Create a dictionary mapping Pokémon types to numerical IDs
type_ids = range(len(types))
type_labels = dict(zip(types, type_ids))

# Prepare images and labels for model training
images = []
labels = []

for file in file_list:
    img = cv2.imread(os.path.join(directory, file))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
    img = cv2.resize(img, (120, 120))  # Resize
    label = type_labels[pokemon_dict[file.split(".")[0]]]
    images.append(img)
    labels.append(label)

images = np.array(images, dtype=np.float32) / 255.0
labels = np.array(labels, dtype=np.int8).reshape(-1, 1)

# Define the model architecture
model = Sequential([
    Flatten(input_shape=(120, 120, 3)),
    Dense(100, activation='relu'),
    Dense(100, activation='relu'),
    Dense(100, activation='relu'),
    Dense(len(types))  # Number of output classes
])

# Print model summary
model.summary()

# Compile the model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Train the model
history = model.fit(images, labels, epochs=50)

def predict_pokemon(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
    img = cv2.resize(img, (120, 120))  # Resize
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    img = img / 255.0  # Normalize
    prediction = model.predict(img)
    predicted_type_id = np.argmax(prediction)
    predicted_type = types[predicted_type_id]

    # Find the Pokémon name corresponding to the predicted type
    predicted_name = None
    for name, type in pokemon_dict.items():
        if type == predicted_type:
            predicted_name = name
            break

    return predicted_name

# Example usage
image_path = './images/charizard.png'
predicted_pokemon = predict_pokemon(image_path)
print("Predicted Pokémon:", predicted_pokemon)

# Loop over file_list to calculate accuracy of the model
correct_predictions = 0
total_images = len(file_list)

for file in file_list:
    image_path = os.path.join(directory, file)
    predicted_pokemon = predict_pokemon(image_path)
    actual_pokemon = file.split('.')[0]
    if predicted_pokemon == actual_pokemon:
        correct_predictions += 1

accuracy = correct_predictions / total_images * 100 
print("Model accuracy:", accuracy)
```
